# EDA - Leisure Centre Dataset

This notebook contains a simple exploratory data analysis of the synthetic leisure centre customer behaviour dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('data/leisure_centre_dataset.csv')
df.head()
```

```python
# Check the structure of the dataset
df.info()
```

```python
# Summary statistics for numerical columns
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Distribution of visits per month
df['visits_per_month'].hist()
plt.title('Visits per Month Distribution')
plt.xlabel('Visits per Month')
plt.ylabel('Frequency')
plt.show()
```
